```python
#binning->converting contin values to a standardized value
#required when creating graphs ->convert numeric cols to discrete variable

pd.cut(data['Age'],bins=[17,30,45,61]) #bin1=17-30,bin2=30-45,...
```

```python
pd.cut(data['Age'],bins=[17,30,45,61],labels=['youth','mid-age','adult']) #no.of labels is 1 less than no of bins specified
```

```python
data['Age_cat']=pd.cut(data['Age'],bins=[17,30,45,61],labels=['youth','mid-age','adult'],ordered=False)
```

```python
data['Sal']=pd.cut(data['EstimatedSalary'],bins=[10000,20000,30000,35000,40000],labels=[''])
```

```python
#pivot table
#col should be categorical in nature
pd.pivot_table(data,columns='Age_cat',values='EstimatedSalary',index='Purchased',aggfunc=(np.mean))
```

```python
#encoding
#categorical encoding(or label encoding) , OHE
#label->numerical vals to a categorical values  according to hierarchy
data['Age_cat'].cat.codes
```

```python
df=pd.get_dummies(data['Age_cat']) #OHE
```
