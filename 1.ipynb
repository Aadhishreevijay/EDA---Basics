```python
'''2. Use pandas and matplotlib/seaborn for the following
a. Read the stores.xlsx file (file given in LMS)
b. Create a column called “shipment_days” which should contain the difference in
days between order date and ship date
c. Create suitable graphs using the numeric columns available in the data by using
matplotlib or seaborn.
'''
```

```python
df=pd.read_excel("stores.xlsx")
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df['shipment_days']=(df['Ship Date']-df['Order Date']).dt.days
```

```python
df.head()
```

```python
sns.histplot(data=df, x='shipment_days', bins=20)
```

```python
sns.histplot(data=df, x='Profit', bins=20)
```

```python
sns.histplot(data=df, x='Sales', bins=20)
```

```python
sns.histplot(data=df, x='Discount', bins=20)
```

```python
sns.histplot(data=df, x='Quantity', bins=20)
```

```python
sns.scatterplot(x='shipment_days', y='Sales', data=df)
```
