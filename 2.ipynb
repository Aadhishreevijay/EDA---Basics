```python
col=['occupation','age']
df=data1[col].groupby('occupation')
Min=df.min().rename(columns={'age':'Min Age'})
Max=df.max().rename(columns={'age':'Max Age'})
com=pd.concat((Min,Max),axis=1)
com
```
