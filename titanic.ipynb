```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Cabin'] = data['Cabin'].ffill()
data['Embarked'] = data['Embarked'].ffill()
```

```python
data.info()
```

```python
sns.pairplot(data)
```

```python
sns.histplot(data=data,x='Fare',bins=20)
```

```python
sns.barplot(data=data,x='Fare',y='Sex')
```

```python
number=data.select_dtypes(exclude=['object'])
number.describe()
```

```python
object=data.select_dtypes(include=['object'])
for i in object:
    print(f"distribution of {i}\n")
    print(data[i].value_counts())
```
