```python
#1. Read the dataset as a pandas dataframe.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('Marketing Analytics.csv')
```

```python
data.head(10)
```

```python
data.info()
```

```python
#2. Identify features that contain outliers by plotting them
```

```python
feature=['Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts','MntFishProducts', 'MntSweetProducts', 'MntGoldProds','NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases','NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue', 'Response', 'Age', 'Customer_Days','marital_Divorced', 'marital_Married','marital_Single','marital_Together', 'marital_Widow', 'education_2n Cycle', 'education_Basic','education_Graduation', 'education_Master', 'education_PhD', 'MntTotal', 'MntRegularProds','AcceptedCmpOverall']
for i in feature:
    plt.boxplot(data[i])
    plt.title(f'Boxplot of {i}')
    plt.show()
```

```python
#3. Remove outliers in the Income column and replace the null values with the mean
```

```python
data['Income']=data['Income'].fillna(data['Income'].mean())
q1=data['Income'].quantile(0.25)
q3=data['Income'].quantile(0.75)
iqr=q3-q1
data=data[(data['Income']>=q1-1.5*iqr)&(data['Income']<= q3+1.5*iqr)]
```

```python
data.info()
```

```python
#4. Create a new column and add the total of the amount spent
```

```python
data['Total_spent']=data[['MntWines', 'MntFruits', 'MntMeatProducts','MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
data.head()
```

```python
#5. create a new column with the total of purchases from the different channels (web,catalogue, instore, deals)
```

```python
data['TotalPurchases']=data[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumDealsPurchases']].sum(axis=1)
data.head()
```

```python
#6. Find whether there exists any relationship between age and income using visualization
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Age'],data['Income'],alpha=0.5,color='Teal')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income')
plt.show()
```

```python
#7. Remove the unwanted columns and save it as a new dataset
```

```python
unwanted=["AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","AcceptedCmp2","AcceptedCmp1","Complain","Z_CostContact","Z_Revenue","Response","Age","Customer_Days"]
cleandata=data.drop(columns=unwanted)
cleandata.head()
```

```python
#1. Read the datasets a pandas dataframe.#
```

```python
d1=pd.read_csv('California Housing Data 1.csv')
d2=pd.read_csv('California Housing Data 2.csv')
```

```python
d1.head()
```

```python
d2.head()
```

```python
d1.info()
```

```python
d2.info()
```

```python
#2. Merge the two dataframes into a single dataframe. Try all 4 types of joins and store the results as separate dataframes.
```

```python
c# Inner join
d_inner=pd.merge(d1,d2,on='ID',how='inner')
# Left join
d_left=pd.merge(d1,d2,on='ID',how='left')
# Right join
d_right=pd.merge(d1,d2,on='ID',how='right')
# Outer join
d_outer=pd.merge(d1,d2,on='ID',how='outer')
print("Inner Join:")
print(d_inner)
print("\nLeft Join:")
print(d_left)
print("\nRight Join:")
print(d_right)
print("\nOuter Join:")
print(d_outer)
```

```python
#3. Remove the duplicate values from ID column.
```

```python
d=d_inner
d.drop_duplicates(subset='ID',keep='first',inplace=True)
d.head()
```

```python
#4. Drop the latitude and longitude columns.
```

```python
d=d.drop(['latitude','longitude'],axis=1)
d.head()
```

```python
#5. Plot histograms of any 5 features
```

```python
feature=['total_rooms','total_bedrooms','median_income','median_income','population']
for i in feature:
    plt.figure()
    d[i].plot(kind='hist',bins=10,edgecolor='black',facecolor='teal')
    plt.title(f'Histogram of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#6. Create a scatter plot and check if there is any relationship between the housing_median_age and median_house_value.
```

```python
plt.figure(figsize=(13,13))
plt.scatter(d['housing_median_age'], d['median_house_value'],alpha=0.5,color='Teal')
plt.title('Scatter Plot: housing median age vs median house value')
plt.xlabel('housing median age')
plt.ylabel('median house value')
plt.show()
```

```python
#7. Perform bivariate analysis for these two columns
```

```python
correlation_coeff=d['housing_median_age'].corr(d['median_house_value'])
print(f"Correlation coefficient: {correlation_coeff}")
plt.figure(figsize=(12,10))
sns.regplot(x='housing_median_age',y='median_house_value',data=d,scatter_kws={'color':'teal'},line_kws={'color':'black'})
plt.title('Scatter Plot: housing_median_age vs median_house_value')
plt.xlabel('housing median age')
plt.ylabel('median house value')
plt.show()
```
